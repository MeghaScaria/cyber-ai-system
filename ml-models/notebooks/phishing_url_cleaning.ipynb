## Data Preprocessing Pipeline - Phishing URLs

#### Dataset Loading, Intial Dataset Inspection and Missing Value Check

```python
import pandas as pd

# Load dataset
df = pd.read_csv("datasets/webpage_url_phishing_dataset.csv")

# Show basic info
print("Dataset Info:")
print(df.info())

# Check missing values
print("\nMissing Values Per Column:")
print(df.isnull().sum())

# Total missing values
print("\nTotal Missing Values:")
print(df.isnull().sum().sum())
```

#### Duplicate Record Detection

```python
import pandas as pd

df = pd.read_csv("datasets/webpage_url_phishing_dataset.csv")

duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)
```

```python
print(df["status"].value_counts())
```

```python
print(df.info())
```

```python
df.drop(columns=["url"], inplace=True)
```

#### Target Label Encoding

```python
df["status"] = df["status"].map({
    "legitimate": 0,
    "phishing": 1
})
```

#### Correlation Analysis

```python
pd.set_option('display.max_rows', None)

correlation = df.corr(numeric_only=True)
print(correlation["status"].sort_values(ascending=False).to_string())
```

#### Feature Selection

```python
selected_features = [
    "google_index",
    "ratio_digits_url",
    "domain_in_title",
    "phish_hints",
    "ip",
    "nb_qm",
    "length_url",
    "nb_slash",
    "length_hostname",
    "nb_eq",
    "ratio_digits_host",
    "shortest_word_host",
    "prefix_suffix",
    "longest_word_path",
    "tld_in_subdomain"
]
```

#### Removing Duplicate Values, Final Dataset Creation and Saving Final Cleaned Dataset

```python
# Step 1: Remove duplicates first
df.drop_duplicates(inplace=True)

# Step 2: Select only chosen features + target label
df_selected = df[selected_features + ["status"]]

# Step 3: Save final cleaned dataset
df_selected.to_csv(
    "datasets/phishing_urls_selected_features.csv",
    index=False
)

print("Final cleaned phishing dataset saved successfully.")
```
